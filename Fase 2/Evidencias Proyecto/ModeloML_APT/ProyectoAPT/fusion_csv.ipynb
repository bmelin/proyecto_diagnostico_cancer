```python
# ============================================
# FUSIÓN DE CSV Y ELIMINACIÓN DE DUPLICADOS
# ============================================

import pandas as pd
import glob
import os

print("Librerías cargadas correctamente.")
```

```python
# ============================================
# CONFIGURACIÓN DE DIRECTORIO
# ============================================

# Ruta donde están los CSV (puedes cambiarla)
carpeta_csv = "datasets/"

# Buscar todos los archivos CSV en la carpeta
archivos = glob.glob(os.path.join(carpeta_csv, "*.csv"))

print(f"Se encontraron {len(archivos)} archivos CSV:")
for a in archivos:
    print("-", os.path.basename(a))
```

```python
# ============================================
# UNIR TODOS LOS CSV EN UN SOLO DATAFRAME
# ============================================

# Leer y concatenar todos los archivos
df_list = [pd.read_csv(a) for a in archivos]
df_final = pd.concat(df_list, ignore_index=True)

print(f"Total de registros combinados: {len(df_final)}")
print(f"Total de columnas: {len(df_final.columns)}")
display(df_final.head())
```

```python
# ============================================
# ELIMINAR DATOS DUPLICADOS
# ============================================

# Si tienes una columna única (por ejemplo 'rut'), usa esta línea:
if 'rut' in df_final.columns:
    df_final = df_final.drop_duplicates(subset=['rut'], keep='first')
else:
    # Si no hay una columna identificadora, elimina duplicados por fila completa
    df_final = df_final.drop_duplicates(keep='first')

print(f"Total de registros después de eliminar duplicados: {len(df_final)}")
```

```python
# ============================================
# GUARDAR CSV FINAL
# ============================================

output_path = "pacientes_10000_final.csv"
df_final.to_csv(output_path, index=False)

print(f"Archivo final guardado como '{output_path}'")
print(f"Tamaño final: {len(df_final)} registros únicos.")
```

```python
# ============================================
# VERIFICAR SI AÚN QUEDAN DUPLICADOS
# ============================================

if 'rut' in df_final.columns:
    duplicados = df_final[df_final.duplicated(subset=['rut'], keep=False)]
else:
    duplicados = df_final[df_final.duplicated(keep=False)]

if len(duplicados) > 0:
    print(f"Aún hay {len(duplicados)} duplicados.")
    display(duplicados)
else:
    print("No se encontraron duplicados en el CSV final.")
```
